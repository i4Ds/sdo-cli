```python
import wandb

api = wandb.Api()
runs = api.runs("mariusgiger/sdo-sood")
```

```python
for run in runs:
    print(f"run {run.id} with state {run.state}")
```

```python
# delete invalid runs
for run in runs:
    if run.state == "failed" or run.state == "crashed":
        run.delete(delete_artifacts=True)
```
